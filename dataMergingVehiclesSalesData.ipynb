```python
import pandas as pd
import re
from IPython.display import display

car_prices_old = pd.read_csv('car_prices_old.csv')
newData = pd.read_csv('newData (1).csv')

# standardize column names
car_prices_old.columns = car_prices_old.columns.str.strip().str.lower()
newData.columns = newData.columns.str.strip().str.lower()

# normalize text so matching is cleaner
def normalize_text(x):
    if pd.isna(x):
        return ""
    x = str(x).lower().strip()
    x = re.sub(r"[-_/]", " ", x)
    x = re.sub(r"[^a-z0-9\s]", "", x)
    x = re.sub(r"\s+", " ", x).strip()
    return x

car_prices_old["make_std"] = car_prices_old["make"].apply(normalize_text)
car_prices_old["model_std"] = car_prices_old["model"].apply(normalize_text)

newData["make_std"] = newData["make"].apply(normalize_text)
newData["model_std"] = newData["model"].apply(normalize_text)

car_prices_old["year"] = pd.to_numeric(car_prices_old["year"], errors="coerce")
newData["year"] = pd.to_numeric(newData["year"], errors="coerce")

# merge only when make, model, year are the same
merged_df = car_prices_old.merge(
    newData,
    on=["make_std", "model_std", "year"],
    how="inner",
    suffixes=("_old", "_new")
)

print("car_prices_old shape:", car_prices_old.shape)
print("newData shape:", newData.shape)
print("merged_df shape:", merged_df.shape)

display(merged_df.head())

```
